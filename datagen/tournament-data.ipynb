```python
from webbrowser import Chrome

import pandas as pd
import requests
from bs4 import BeautifulSoup
from requests import get
from selenium import webdriver
from time import sleep
```

```python
events = [1611,7524,6372,6586,1333,2027,1444,4866,1617,2471,1666,4443,1553,6793,3564,3247,2062,2720,3883,7148]
```

```python
matches = []

for event_id in events:
    dr = webdriver.Chrome()
    dr.get(f"https://www.hltv.org/results?event={event_id}")
    sleep(0.2)
    print('retreived')
    bs = BeautifulSoup(dr.page_source)
    dr.close()

    event_field= bs.find('div', attrs={'class': 'event-hub-title'})

    event = event_field.text
    for day in  bs.find_all('div', attrs={'class': 'results-sublist'}):
        date = day.find('div', attrs={'class': 'standard-headline'}).text
        for matchup in day.find_all('a'):
            stats = matchup.attrs['href']

            loser_score = int(matchup.find('span', attrs={'class': 'score-lost'}).text)
            winner_score = int(matchup.find('span', attrs={'class': 'score-won'}).text)
            teams = [t.text for t in matchup.find_all('div', attrs={'class': 'team'})]
            winner = matchup.find('div', attrs={'class': 'team team-won'}).text
            teams.remove(winner)
            loser = teams[0]
            matches.append({'date': date, 'event': event, 'event-id': event_id, 'winner': winner, 'loser': loser, 'winner-score': winner_score, 'loser-score': loser_score, 'stats': 'https://www.hltv.org' + stats})

matches
```

```python
matchups = pd.DataFrame(matches)
matchups.to_parquet('../matches.parquet')
matchups
```
